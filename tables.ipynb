```python
import pandas as pd
import numpy as np
```

```python
cov = [[2**2,0.4*2*2], [0.4*2*2,2**2]]
mean = [7.5,7.5]
```

```python
x = np.random.multivariate_normal(mean, cov, 15)
```

```python
def mcar(x):
    if(np.random.uniform(0,1)<0.5):
        return 0.0
    else:
        return x
```

```python
def mar(x, y):
    if(x<7.5):
        return y
    else:
        return 0.0
```

```python
def mnar(y):
    if(y<7.5):
        return y
    else:
        return 0.0
```

```python
df = pd.DataFrame(x, columns = ["2018", "2019 (Complete)"])
df.index = np.arange(1,16)
df["2019 (MCAR)"] = df["2018"].map(lambda x: mcar(x))
df["2019 (MAR)"] = df.apply(lambda x: mar(x["2018"], x["2019 (Complete)"]), axis=1)
df["2019 (MNAR)"] = df.apply(lambda x: mnar(x["2019 (Complete)"]), axis=1)
```
